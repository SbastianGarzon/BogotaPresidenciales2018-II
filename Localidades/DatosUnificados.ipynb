```python
%pylab inline
import pandas as pd
import os 
```

Construcción base de datos:

```python
#Inicializo las columnas con los nombres de las localidades que ya tienen datos
localidades=os.listdir("Votaciones_localidades")
Nombreslocalidades=[]
for i in localidades:
    Nombreslocalidades.append(i[:-4])
df = pd.DataFrame(columns=[Nombreslocalidades])
df= df.rename_axis(None)

```

```python
#Algoritmo que ingresa los datos a la tabla, la tabla solo contiene las localidades reportadas.
partidos=('Gustavo Petro','Ivan Duque','Potencial Votantes','Blanco','Nulos','No Marcados','Referencia:')
for i in localidades:
    temp=pd.read_csv("Votaciones_localidades/"+str(i))
    temp.index = partidos
    df[str(i[:-4])] = temp.iloc[:-1,1]
df2=df.apply(pd.to_numeric).copy()
```

## Base de datos completa

```python
#Suma el total de cada Item en Bogotá y genera un archivo CSV con la información
dc=df2.copy()

dc["Total Bogota"]=dc.sum(axis=1)
total = dc.sum()-dc.loc['Potencial Votantes']
total.name="Total votantes"
porcentaje=(total/dc.loc["Potencial Votantes"])*100
porcentaje.name="Porcentaje votantes(%)"
dfinal=dc.append([total,porcentaje]).copy()
formato= lambda x: format(x,'.1f')
dfinal.loc["Porcentaje votantes(%)"]=dfinal.loc['Porcentaje votantes(%)'].apply(formato)
dfinal.to_csv("../Bogota/Votacion_Bogota_Unificada.csv")
```
